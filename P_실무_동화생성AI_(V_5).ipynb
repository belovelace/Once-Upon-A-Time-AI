```python
import json

# ==============================================================================
# [Helper] ì ìˆ˜í‘œ ì¶œë ¥ í•¨ìˆ˜ (ì‹œê°ì  íš¨ê³¼)
# ==============================================================================
def print_score_card(attempt, result):
    scores = result.get('scores', {})
    reasons = result.get('reasons', {})
    decision = result.get('final_decision', 'UNKNOWN')

    print(f"\nğŸ“Š [ê²€ì¦ ë¦¬í¬íŠ¸ - ì‹œë„ {attempt}íšŒì°¨]")
    print("-" * 50)
    print(f"1. ìœ í•´ì„± (Safety)    : {'â˜…' * scores.get('safety', 0)} ({scores.get('safety', 0)}/5)")
    print(f"2. ì €ì‘ê¶Œ (Copyright) : {'â˜…' * scores.get('copyright', 0)} ({scores.get('copyright', 0)}/5)")
    print(f"3. êµí›ˆì„± (Alignment) : {'â˜…' * scores.get('alignment', 0)} ({scores.get('alignment', 0)}/5)")
    print(f"4. ì‹¬ë¯¸ì„± (Aesthetics): {'â˜…' * scores.get('aesthetics', 0)} ({scores.get('aesthetics', 0)}/5)")
    print("-" * 50)
    print(f"ğŸ§ ì¢…í•© íŒì •: [{'âœ… PASS' if decision == 'PASS' else 'âŒ FAIL'}]")

    if decision == 'FAIL':
        # Display all relevant reasons if the decision is FAIL
        fail_reasons = []
        for key, value in scores.items():
            if value < 3: # Assuming 3 is the minimum passing score for safety/copyright
                if key == 'safety' or key == 'copyright': # Critical failures
                    fail_reasons.append(reasons.get(key, ''))
            elif key == 'alignment' or key == 'aesthetics': # Warnings for quality scores below 3
                if value < 3:
                    fail_reasons.append(reasons.get(key, ''))

        if not fail_reasons: # Fallback if no specific reason found
            fail_reasons.append('ê¸°ì¤€ ë¯¸ë‹¬')

        print(f"âš ï¸ ì‚¬ìœ : {', '.join([r for r in fail_reasons if r])}")
    print("=" * 60)


# ==============================================================================
# [ë©”ì¸ ì‹¤í–‰ê¸°] ì „ì²´ í”„ë¡œì„¸ìŠ¤ í†µí•©
# ==============================================================================
def run_process_with_display(user_input):
    max_retries = 3
    attempt = 0
    current_story = ""
    story_plan = {}
    feedback = None
    is_finalized = False
    initial_story_and_plan_set = False # New flag to track if initial story and plan are successfully generated

    print(f"\nğŸš€ [System] ë™í™” ìƒì„± í”„ë¡œì„¸ìŠ¤ë¥¼ ì‹œì‘í•©ë‹ˆë‹¤: ì£¼ì œ '{user_input}'")

    # -------------------------------------------------------
    # ìƒì„± ë° ê²€ì¦ ë°˜ë³µ ë£¨í”„ (Retry Loop)
    # -------------------------------------------------------
    while attempt < max_retries:
        attempt += 1

        # 1. ìŠ¤í† ë¦¬ ìƒì„± (ì´ˆì•ˆ vs ì¬ì‘ì„±ì•ˆ)
        if not initial_story_and_plan_set: # First attempt to generate story and plan
            print(f"\nğŸ“ [Step 1] ì´ˆì•ˆ ì‘ì„± ì¤‘... (Attempt {attempt})")
            generated_story, generated_plan = generate_fairy_tale(user_input)

            if generated_story is None or generated_plan is None: # Handle safety check or other initial generation failures
                print("âŒ [System] ì´ˆê¸° ë™í™” ìƒì„±ì— ì‹¤íŒ¨í–ˆìŠµë‹ˆë‹¤. ì‚¬ìš©ì ì…ë ¥ì´ ë¶€ì ì ˆí•˜ê±°ë‚˜ ì‹œìŠ¤í…œ ì˜¤ë¥˜ê°€ ë°œìƒí–ˆìŠµë‹ˆë‹¤.")
                return None # Exit if initial generation fails

            current_story = generated_story
            story_plan = generated_plan
            initial_story_and_plan_set = True # Initial plan and story are set

            print("\n" + "â”Œ" + "â”€"*58 + "â”")
            print("â”‚  ğŸ“œ [ì´ˆì•ˆ (Draft Version)]                           â”‚")
            print("â””" + "â”€"*58 + "â”˜")
            print(current_story)
            print("-" * 60)

        else: # Subsequent attempts for refinement using the existing story_plan
            print(f"\nğŸ”„ [Step 1-Retry] í”¼ë“œë°± ë°˜ì˜í•˜ì—¬ ì¬ì‘ì„± ì¤‘... (Attempt {attempt})")
            print(f"   (ë°˜ì˜í•  í”¼ë“œë°±: {feedback})")
            # Use refine_story for rewrites, passing the current_story, feedback, and original story_plan
            current_story = refine_story(current_story, feedback, story_plan)

            print("\n" + "â”Œ" + "â”€"*58 + "â”")
            print(f"â”‚  ğŸ“œ [ìˆ˜ì •ì•ˆ {attempt} (Revised Draft)]                       â”‚")
            print("â””" + "â”€"*58 + "â”˜")
            print(current_story)
            print("-" * 60)

        # 2. ê²€ì¦ ìˆ˜í–‰ (Step 2)
        print("\nğŸ” [Step 2] í’ˆì§ˆ ê²€ì¦ ìˆ˜í–‰ ì¤‘...")
        verification_result = verify_story_quality(current_story, story_plan)

        # 3. ì ìˆ˜í‘œ ì¶œë ¥ (ìš”ì²­í•˜ì‹  ë¶€ë¶„)
        print_score_card(attempt, verification_result)

        # 4. ê²°ê³¼ì— ë”°ë¥¸ ë¶„ê¸° ì²˜ë¦¬
        decision = verification_result.get('final_decision', 'FAIL')

        if decision == "PASS":
            # (ì˜µì…˜) ë¬¸ì²´ ìŠ¤íƒ€ì¼ êµì • - 'refine_style' í•¨ìˆ˜ê°€ ì •ì˜ë˜ì–´ ìˆì§€ ì•Šìœ¼ë¯€ë¡œ ì œê±°í•©ë‹ˆë‹¤.
            # scores = verification_result.get('scores', {})
            # if scores.get('aesthetics', 0) < 4: # 4ì  ë¯¸ë§Œì´ë©´ ìŠ¤íƒ€ì¼ ë‹¤ë“¬ê¸°
            #     print("\nâœ¨ [Step 3] ë‚´ìš©ì€ í†µê³¼í–ˆìœ¼ë‚˜, ë¬¸ì²´ë¥¼ ë” ë¶€ë“œëŸ½ê²Œ ë‹¤ë“¬ìŠµë‹ˆë‹¤ (Polishing)...")
            #     current_story = refine_style(current_story)
            # else:
            print("\nâœ… [Step 3] ìˆ˜ì • ì—†ì´ ìµœì¢…ì•ˆìœ¼ë¡œ í™•ì •í•©ë‹ˆë‹¤.")

            is_finalized = True
            break

        else:
            # FAIL ì²˜ë¦¬
            feedback = verification_result.get('rewrite_suggestion', 'ë‚´ìš© ìˆ˜ì • í•„ìš”')
            if attempt < max_retries:
                print(f"ğŸš« [System] ê¸°ì¤€ ë¯¸ë‹¬ë¡œ ë‹¤ì‹œ ì‘ì„±í•©ë‹ˆë‹¤.")
            else:
                print("âŒ [System] ìˆ˜ì • í•œë„ë¥¼ ì´ˆê³¼í–ˆìŠµë‹ˆë‹¤. ìµœì¢… ìƒì„±ì— ì‹¤íŒ¨í–ˆìŠµë‹ˆë‹¤.")

    # -------------------------------------------------------
    # ìµœì¢… ê²°ê³¼ ì¶œë ¥ (Final Output)
    # -------------------------------------------------------
    if is_finalized:
        print("\n" + "="*60)
        print("ğŸ‰ [Final Result] ìµœì¢… ì™„ì„±ëœ ë™í™”")
        print("="*60)
        print(current_story)
        print("="*60)

        # JSON ë³€í™˜
        final_json = format_final_output(current_story, story_plan)
        return final_json
    else:
        print("\nâš ï¸ ìµœì¢… ìƒì„± ì‹¤íŒ¨.")
        return None
```
